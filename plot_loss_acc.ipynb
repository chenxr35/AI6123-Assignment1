```python
import numpy as np
import matplotlib.pyplot as plt
import os

%load_ext autoreload
%autoreload 2
```

```python
def load_npz(file_path):
    save_data = np.load(file_path, allow_pickle=True)
    return save_data['train_loss'],save_data['valid_loss'],save_data['train_acc'],save_data['valid_acc']
```

```python
def plot_train_loss(npz_files, labels, fig_title):
    plt.figure(figsize=(10, 6))
    for i, file_path in enumerate(npz_files):
        train_loss, _, _, _ = load_npz(file_path)
        epochs = np.arange(1, len(train_loss) + 1)
        plt.plot(epochs, train_loss, '-.', label=f'{labels[i]}', linewidth=1.5)

    plt.title(fig_title)
    plt.xlabel('Epoch')
    plt.ylabel('Training Loss')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
def plot_valid_loss(npz_files, labels, fig_title):
    plt.figure(figsize=(10, 6))
    for i, file_path in enumerate(npz_files):
        _, valid_loss, _, _ = load_npz(file_path)
        epochs = np.arange(1, len(valid_loss) + 1)
        plt.plot(epochs, valid_loss, '-.', label=f'{labels[i]}', linewidth=1.5)

    plt.title(fig_title)
    plt.xlabel('Epoch')
    plt.ylabel('Validation Loss')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
def plot_loss(npz_files, labels, fig_title):
    plt.figure(figsize=(10, 6))
    for i, file_path in enumerate(npz_files):
        train_loss, valid_loss, _, _ = load_npz(file_path)
        epochs = np.arange(1, len(train_loss) + 1)
        plt.plot(epochs, train_loss, '-.', label=f'{labels[i]} training', linewidth=1.5)
        plt.plot(epochs, valid_loss, '-.', label=f'{labels[i]} validation', linewidth=1.5)

    plt.title(fig_title)
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
def plot_train_acc(npz_files, labels, fig_title):
    plt.figure(figsize=(10, 6))
    for i, file_path in enumerate(npz_files):
        _, _, train_acc, _ = load_npz(file_path)
        epochs = np.arange(1, len(train_acc) + 1)
        plt.plot(epochs, train_acc, '-.', label=f'{labels[i]}', linewidth=1.5)

    plt.title(fig_title)
    plt.xlabel('Epoch')
    plt.ylabel('Training Accuracy')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
def plot_valid_acc(npz_files, labels, fig_title):
    plt.figure(figsize=(10, 6))
    for i, file_path in enumerate(npz_files):
        _, _, _, valid_acc = load_npz(file_path)
        epochs = np.arange(1, len(valid_acc) + 1)
        plt.plot(epochs, valid_acc, '-.', label=f'{labels[i]}', linewidth=1.5)

    plt.title(fig_title)
    plt.xlabel('Epoch')
    plt.ylabel('Validation Accuracy')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
def plot_acc(npz_files, labels, fig_title):
    plt.figure(figsize=(10, 6))
    for i, file_path in enumerate(npz_files):
        _, _, train_acc, valid_acc = load_npz(file_path)
        epochs = np.arange(1, len(train_acc) + 1)
        plt.plot(epochs, train_acc, '-.', label=f'{labels[i]} training', linewidth=1.5)
        plt.plot(epochs, valid_acc, '-.', label=f'{labels[i]} validation', linewidth=1.5)

    plt.title(fig_title)
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
npz_files = ['RNN_SGD_epochs20_pretrainedFalse.npz', 
             'RNN_Adagrad_epochs20_pretrainedFalse.npz',
             'RNN_Adam_epochs20_pretrainedFalse.npz']
labels = ['SGD', 'Adagrad', 'Adam']  
fig_title = "Experiments with different optimizers"
```

```python
plot_loss(npz_files, labels, fig_title)
```

```python
plot_acc(npz_files, labels, fig_title)
```

```python
npz_files = ['RNN_Adam_epochs50_pretrainedFalse.npz',
             'OneLayer_FNN_Adam_epochs50_pretrainedFalse.npz',
             'TwoLayer_FNN_Adam_epochs50_pretrainedFalse.npz',
             'ThreeLayer_FNN_Adam_epochs50_pretrainedFalse.npz',
             'CNN_Adam_epochs50_pretrainedFalse.npz',
             'LSTM_Adam_epochs50_pretrainedFalse.npz',
             'BiLSTM_Adam_epochs50_pretrainedFalse.npz']
labels = ['RNN', 'One-layer FNN', 'Two-layer FNN', 'Three-layer FNN', 'CNN', 'LSTM', 'Bi-LSTM']  
fig_title = "Experiments with different models"
```
